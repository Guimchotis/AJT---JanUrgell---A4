```python
def fibonacci(n):
    a, b = 0, 1
    fib = []
    for _ in range(n + 1):
        fib.append(a)
        a, b = b, a + b
    return fib
```

```python
print (fibonacci(29))
```

```python
def mostra_taulell(taulell, diccionari_jugadors, N):
    print("    " + "   ".join(str(i) for i in range(N)))
    for fila in range(N):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(N))
        print(f" {fila}  {contingut}")
        if fila < N - 1: print("   " + "---+" * (N - 1) + "---")

def demana_posicio(N):
    try:
        f = int(input(f"Fila (0-{N-1}): "))
        c = int(input(f"Columna (0-{N-1}): "))
        return (f, c) if 0 <= f < N and 0 <= c < N else (None, None)
    except: return None, None

def comprovar_victoria(taulell, jugador, N):
    for i in range(N):
        if all(taulell[i][j] == jugador for j in range(N)) or all(taulell[j][i] == jugador for j in range(N)): return True
    return all(taulell[i][i] == jugador for i in range(N)) or all(taulell[i][N-1-i] == jugador for i in range(N))


jugador, diccionari_jugadors, comptador, condicio_victoria = 1, {0: " ", 1: "X", -1: "O"}, 0, False
N = int(input("Mida N: "))
taulell = [[0] * N for _ in range(N)] 

while not condicio_victoria and comptador < N**2:
    clear_output(wait=True)
    mostra_taulell(taulell, diccionari_jugadors, N)
    fila, columna = demana_posicio(N)
    
    if fila is None: break
    if taulell[fila][columna] != 0: continue 

    taulell[fila][columna] = jugador
    condicio_victoria = comprovar_victoria(taulell, jugador, N)
    
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell, diccionari_jugadors, N)
        print(f"Guanya {diccionari_jugadors[jugador]}")
        break
    
    jugador, comptador = -jugador, comptador + 1 
```

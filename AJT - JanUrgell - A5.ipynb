```python
class Personatge:
    def __init__(self, nom, vida, força, nivell = 1):
        
        self.nom = nom
        self.vida = vida
        self.força = força
        self.nivell = nivell

    def atacar(self, castor):
        
        print(f"{self.nom} mossega a {castor.nom}")
        castor.vida -= self.força
        if castor.vida < 0:
            castor.vida = 0

    def pujar_nivell(self):
        
        self.nivell += 1
        self.força += 2
        self.vida += 5

    def esta_viu(self):
       
        return self.vida > 0

    def info(self):
        
        return f"El {self.nom} ha derrotat al seu rival d'una mossegada i amb {self.vida} punts de vida sobrants. Amb una força de {self.força}. A més ha adquirit un nivell de {self.nivell}."

    def __str__(self):
       
        return self.info()



Coletas = Personatge("Coletas", 69, 15)
Castor = Personatge("Castor", 100, 10)


while Coletas.esta_viu() and Castor.esta_viu():
    
    Coletas.atacar(Castor)
    if not Castor.esta_viu():
        guanyador = Coletas
        break
        
    
    Castor.atacar(Coletas)
    if not Coletas.esta_viu():
        guanyador = Castor
        break


print("\n\n\n||||||Fi de l'enfrontament||||||\n")
guanyador.pujar_nivell()
print(guanyador.info())

```
